Question-1 :-  What is the difference between AI, ML, DL, and Data Science ?Provide a brief explanation of each.

Answer-1 :-
1. Artificial Intelligence (AI) :

Artificial intelligence is the broad ambition. It includes any system designed to perform tasks that normally require human intelligence. Pattern recognition, planning, language understanding, decision making. If a machine behaves as if it has a mind, it sits under this umbrella. The term is old. It goes back to the nineteen fifties, when computers filled rooms and yet people already imagined them thinking.

 - Scope: The grand vision. It covers anything that makes a machine act smart.

 - Technique: It often uses "Rule-Based" logic (If X happens, do Y).

 - Application (Streaming Example): The entire platform interface is an AI system. It "knows" you are logging in from India and shows you Indian content first.
2. Machine Learning (ML) :

Machine learning is a subset of artificial intelligence. The core idea is that the machine improves its performance through data rather than explicit instructions. Instead of telling the computer every tiny rule, you let it discover patterns. A simple example is a model that learns from past weather data to predict tomorrow’s rainfall. It does not understand clouds, it simply internalizes statistical structure.

 - Scope: The ability to learn patterns without being explicitly programmed.

 - Technique: It uses statistical algorithms. You feed it data (e.g., your watch history), and it finds the mathematical relationship.

 - Application (Streaming Example): The Recommendation Engine.
3. Deep Learning (DL) :

Deep learning is a subset of machine learning. It uses architectures called neural networks with many layers. These layers allow the system to learn complex hierarchies of patterns. Early layers capture simple shapes or edges, while deeper ones capture concepts. Image classification is a classic case. A network can look at pixels and eventually learn to decide whether an image contains a cat. This works because the layers learn features automatically instead of relying on hand crafted rules.

 - Scope: The "Brain-like" approach. It handles unstructured data (images, audio, video) that standard ML struggles with.

 - Technique: Neural Networks. Imagine layers of "neurons" that pass information to each other.

 - Application (Streaming Example): Auto-generating Subtitles or Thumbnails.
4. Data Science (DS) :

Data science is the practical craft of extracting insights from data. It is not only about prediction. It includes collecting data, cleaning it, exploring it, visualizing it, modeling it, and communicating the results. A data scientist uses statistics, programming, machine learning, and domain knowledge. The goal is to turn raw data into something that supports decisions. A retail company uses data science to identify seasonal buying patterns. A hospital uses it to understand patient outcomes.

 - Scope: The "Business Logic." It uses AI, ML, and DL as tools to solve a business problem.

 - Technique: Data Cleaning (fixing messy data), Visualization (graphs/charts), and Storytelling.

 - Application (Streaming Example): Deciding which show to produce next.

These four fields are related, but their purposes differ. Artificial intelligence is the vision, machine learning is one powerful method within that vision, deep learning is a specific technique inside machine learning, and data science is the practice of making sense of data in the real world. They often work together, but each has its own territory, its own tools, and its own quirks.

Question-2 :  Explain overfitting and underfitting in ML. How can you detect and prevent them?

Answer-2 :- Overfitting occurs when a machine learning model learns noise and outliers in the training data, performing well on training data but poorly on unseen test data due to high variance. Underfitting happens when the model is too simple to capture underlying patterns, resulting in high bias and poor performance on both training and test data.​

 - Bias-Variance Tradeoff :

Total error in a model decomposes into bias squared, variance, and irreducible error, where high bias leads to underfitting (systematic errors from oversimplification) and high variance causes overfitting (sensitivity to training data fluctuations). The tradeoff requires balancing model complexity: simpler models reduce variance but increase bias, while complex models do the opposite, aiming for a sweet spot that minimizes overall error.​

 - Detection Methods :

Plot learning curves comparing training and validation errors: overfitting shows low training error but high validation error (high variance), while underfitting shows high errors on both (high bias). Use cross-validation to assess generalization by splitting data into folds and averaging performance metrics like accuracy or MSE across them.​

 - Prevention Techniques :
   - Regularization: Add penalties like L1 (Lasso) or L2 (Ridge) to the loss function to constrain model complexity and reduce variance in overfitting cases.
   - Cross-Validation: Employ k-fold cross-validation to tune hyperparameters and select models that perform consistently across data splits, mitigating both bias and variance.
    - Other Strategies: Increase training data to reduce variance, simplify models (e.g., pruning decision trees) for underfitting, or use ensemble methods like bagging/boosting to average predictions and balance the tradeoff.

Question-3 :- How would you handle missing values in a dataset? Explain at least three methods with examples.

Answer-3 :- Handling missing values in machine learning datasets is crucial to avoid bias and ensure model performance; common methods include deletion, imputation, and predictive modeling, selected based on the missing data percentage, pattern (e.g., MCAR, MAR), and data type.​

 - Listwise Deletion :

Remove entire rows or columns with missing values using techniques like dropna() in pandas, preserving data integrity but risking significant sample size reduction if >5-10% data is missing. Example: In a sales dataset with occasional null prices, dropping those rows yields a clean subset for training, suitable when data loss is minimal and missingness is random.​

 - Mean/Median Imputation :

Replace missing values with the mean (for normally distributed data) or median (for skewed data) of the column, a simple statistical method implemented via scikit-learn's SimpleImputer(strategy='mean'). Example: For a dataset of house prices with missing values, impute age with column median (e.g., 25 years), maintaining dataset size but potentially underestimating variance in outliers.​

 - Predictive Modeling Imputation :

Use algorithms like KNN or regression to predict missing values based on other features, capturing relationships (e.g., via scikit-learn's KNNImputer). Example: In a patient dataset, predict missing blood pressure from age, BMI, and cholesterol using a linear model, offering accuracy for MAR data but computationally intensive.​

Choose methods by evaluating impact via cross-validation; deletion suits low missingness, imputation for moderate, and predictive for high-relationship data.

```python
# Question-4 :- What is an imbalanced dataset? Describe two techniques to handle it (theoretical + practical).

# Answer-4 :-
# 1. Definition :

# An imbalanced dataset occurs in a classification problem when the distribution of examples across the known classes is biased or skewed. One class (the majority class) has significantly more samples than the other class (the minority class).

#    - Formal Definition: If class $A$ has $N$ samples and class $B$ has $M$ samples, and $N >> M$, the dataset is imbalanced.
#    - Consequence: Standard machine learning algorithms maximize overall accuracy. Consequently, they tend to ignore the minority class, treating it as noise, resulting in poor recall/sensitivity for the class of interest (e.g., fraud detection, disease diagnosis).
# 2. Techniques to Handle Imbalanced Data :

# There are two primary approaches: Data-level techniques (Resampling) and Algorithm-level techniques (Cost-Sensitive Learning).

# Method A: Resampling Techniques :

# This involves modifying the training dataset to balance the class distribution before training the model.
# 1. Random Undersampling

#  - Mechanism: Randomly removes samples from the majority class until the dataset is balanced.

#  - Advantage: Reduces computational complexity and training time.

#  - Disadvantage: results in the loss of potentially valuable information, which can lead to underfitting.

# 2. Random Oversampling :

#  - Mechanism: Randomly duplicates samples from the minority class.

#  - Advantage: No information is lost.

#  - Disadvantage: Increases the likelihood of overfitting, as the model memorizes the exact replicated samples.

# 3. SMOTE (Synthetic Minority Over-sampling Technique) :
#  - Mechanism: Instead of duplicating existing data, SMOTE generates synthetic samples. It selects a minority sample, identifies its $k$ nearest neighbors, and creates new points along the vector between the sample and its neighbors.
#  - Mathematical Concept:$$x_{new} = x_i + \lambda \times (x_{neighbor} - x_i)$$
#   Where $\lambda$ is a random number between 0 and 1.
#  - Advantage: Reduces the risk of overfitting compared to simple oversampling because it creates plausible variations of the data.

# Python Implementation (SMOTE):
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
import numpy as np

# Assume X and y are the features and target variable
# Creating a synthetic imbalanced dataset for demonstration
X = np.random.rand(1000, 5) # 1000 samples, 5 features
y = np.zeros(1000) # Majority class
y[:50] = 1 # Minority class (50 samples)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y)

# Apply SMOTE to the training set only
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

print(f"Original training set shape: {X_train.shape}, {y_train.shape}")
print(f"Original class distribution (training): {np.bincount(y_train.astype(int))}")
print(f"Resampled training set shape: {X_train_resampled.shape}, {y_train_resampled.shape}")
print(f"Resampled class distribution (training): {np.bincount(y_train_resampled.astype(int))}")

# Method B: Cost-Sensitive Learning (Class Weights) :

#  - This involves modifying the learning algorithm to penalize misclassifications of the minority class more heavily than the majority class.

#  - Mechanism: Assign a higher weight (penalty) to the minority class in the loss function during training.

#  - Logic: If the minority class is 10 times smaller than the majority class, the algorithm treats 1 error on the minority class as equivalent to 10 errors on the majority class.

#  - Advantage: No data is generated or discarded; the original distribution is preserved while forcing the model to prioritize the minority class.
# Python Implementation (Class Weights):
from sklearn.ensemble import RandomForestClassifier

# 'class_weight="balanced"' calculates weights inversely proportional to class frequencies
clf = RandomForestClassifier(class_weight='balanced', random_state=42)

clf.fit(X_train, y_train)
```

Question 5: Why is feature scaling important in ML? Compare Min-Max scaling and Standardization.

Answer-5 :-
Feature scaling is essential in machine learning because it ensures all features contribute equally to the model's learning process. It improves model performance, speeds up convergence in gradient descent, and is crucial for distance-based algorithms like K-Nearest Neighbors (KNN) and Support Vector Machines (SVM), which rely on feature distances for classification or clustering.​

 - Min-Max Scaling :
   - Description: This technique rescales features to a fixed range, typically, by subtracting the minimum value and dividing by the range (max-min).​
   - Impact: It preserves the original distribution's shape but compresses all features into the same scale, preventing dominance from features with large ranges.
   - Use Case: Ideal when the data doesn't follow a Gaussian distribution and when a bounded range is desired, such as image pixel normalization.
   - Effect on Algorithms: Beneficial for distance-based algorithms like KNN and SVM since features contribute equally to distance calculations; also helps gradient descent converge faster by normalizing the input scale.​

 - Standardization (Z-score Normalization) :
   - Description: This technique transforms features to have zero mean and unit variance by subtracting the mean and dividing by the standard deviation.
   - Impact: It centers the data and scales it according to its spread, which is useful when data is normally distributed.
   - Use Case: Preferred for algorithms assuming data follows a Gaussian distribution, such as linear regression, logistic regression, and PCA.
   - Effect on Algorithms: Helps gradient descent converge more efficiently by equalizing feature variances, and enhances performance in algorithms sensitive to feature distribution like SVM.​

In summary, both methods ensure models are not biased by feature magnitude differences. Min-Max scaling confines values within a range, while Standardization makes data zero-centered with unit variance, each benefiting different types of algorithms and data distributions.

Question 6:  Compare Label Encoding and One-Hot Encoding. When would you prefer one over the other?

Answer-6 :
 - Label Encoding and One-Hot Encoding convert categorical variables to numerical formats for machine learning, differing in representation and model impact.​

 - Label Encoding: Assigns unique integers to categories (e.g., Red=0, Green=1, Blue=2), keeping one column with low memory use but implying false ordinal relationships that can bias linear models.​

 - One-Hot Encoding: Creates binary columns per category (e.g., Red: ), expanding dimensionality but avoiding order assumptions, suitable for independent categories.​

 - Key Differences: Label Encoding is compact and assumes hierarchy; One-Hot treats categories equally without order.​

Prefer Label Encoding for ordinal data (e.g., Low/Medium/High ratings) or tree-based models like Random Forests with many categories. Choose One-Hot Encoding for nominal data (e.g., colors, cities) in distance-based algorithms like KNN/SVM or neural networks to prevent misleading hierarchies

```python
# Question 8: Titanic Dataset
# a) Compare the survival rates based on passenger class (Pclass). Which class had the highest survival rate, and why do you think that happened?
# b) Analyze how age (Age) affected survival. Group passengers into children (Age < 18) and adults (Age ≥ 18). Did children have a better chance of survival?
# Dataset: https://github.com/MasteriNeuron/datasets.git
import pandas as pd

# 1. Load the dataset
url = "https://raw.githubusercontent.com/MasteriNeuron/datasets/master/titanic.csv"

try:
    df = pd.read_csv(url)

    # --- Part A: Survival based on Passenger Class (Pclass) ---
    # Group by Pclass and calculate the mean of 'Survived' (0 = No, 1 = Yes)
    pclass_survival = df.groupby('Pclass')['Survived'].mean().sort_values(ascending=False)

    # --- Part B: Survival based on Age (Child vs Adult) ---
    # Drop rows where Age is missing to ensure accurate grouping
    age_df = df.dropna(subset=['Age']).copy()

    # Create a new column 'AgeGroup'
    age_df['AgeGroup'] = age_df['Age'].apply(lambda x: 'Child' if x < 18 else 'Adult')

    # Group by AgeGroup and calculate survival rate
    age_survival = age_df.groupby('AgeGroup')['Survived'].mean()

    # Output the results
    print("--- Survival Rate by Passenger Class ---")
    print(pclass_survival)

    print("\n--- Survival Rate by Age Group ---")
    print(age_survival)

except Exception as e:
    print(f"Error loading data: {e}")

# Analysis & Reasoning :
# A) Passenger Class (Pclass):
#  - Highest Survival Rate: 1st Class (approx. 63%).

#  - Lowest Survival Rate: 3rd Class (approx. 24%).

#  - Reason:

#  - Proximity to Lifeboats: First-class cabins were located on the upper decks (Decks A-E), closest to where the lifeboats were stationed. Third-class passengers were deep inside the ship and had to navigate complex corridors to reach the top.

#  - Priority & Status: During the chaotic evacuation, officers and crew often prioritized wealthy and high-status passengers.

#  - Physical Barriers: In some instances, gates separating the third-class sections from the upper decks were kept locked or were difficult to open, trapping many below deck.

# B) Age (Child vs. Adult)
# Did children have a better chance? Yes.

#  - Data: Children (under 18) had a survival rate of roughly 54%, compared to 38% for adults.

#  - Reason:

# "Women and Children First": This maritime code of conduct (the Birkenhead Drill) was strictly enforced by the officers loading the lifeboats. Children were physically lifted into boats ahead of men, regardless of their class in many cases.
```
