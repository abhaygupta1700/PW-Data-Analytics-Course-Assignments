Question-1 :- What are Type I and Type II errors in hypothesis testing, and how do they impact decision-making ?

Answer-1 :- Imagine a scientist sitting with two boxes. One box contains a boring pebble, the other contains a tiny diamond. The scientist cannot look inside, so they run a statistical test to decide which box they got. That is the vibe of hypothesis testing. The universe does not send a self-addressed note telling you if your guess is correct. It only lets you control the chances of certain mistakes.

 - A Type I error happens when the test declares a discovery when reality contains only the pebble. In formal terms, it occurs when you reject the null hypothesis even though the null hypothesis is true. In everyday life this is like diagnosing a disease in a healthy person. The preset significance level usually called alpha controls the probability of this mistake.

 - A Type II error happens when the test shrugs and says nothing to see here even though the diamond is sitting in the box. In formal terms, it occurs when you fail to reject the null hypothesis even though the alternative hypothesis is true. This is like telling a sick person that they are healthy. The probability of this mistake is called beta. One minus beta is the power of the test which measures its ability to detect real effects.

These errors bend decision making in opposite directions. If you tighten the rules to avoid false discoveries by picking a small alpha then Type I errors fall but Type II errors rise because the test becomes more cautious. If you loosen the rules to catch more true discoveries then Type II errors fall but you risk more Type I errors. Choosing the balance depends on the consequences. A medical trial tends to treat Type I errors as expensive because approving an ineffective drug harms many people. A security screening system tends to treat Type II errors as expensive because missing a real threat matters more than temporarily inconveniencing innocent passengers.

Question-2 :- What is the P-value in hypothesis testing, and how should it be interpreted in the context of the null hypothesis?

Answer-2 :- The P value is a kind of cosmic shrug that measures how surprising your data would be if the null hypothesis were actually true. The null hypothesis is usually the claim that nothing interesting is happening. For example, there is no difference between two treatments or no association between two variables.

A P value is the probability of obtaining results at least as extreme as the ones you observed assuming the null hypothesis holds. When the P value is small, it means your data would be very unusual under a world where the null hypothesis is correct. This does not prove the null hypothesis is false. It simply says that the data are awkward to explain using the null hypothesis alone.

When the P value is large, your data look comfortably normal under the null hypothesis. It does not prove the null hypothesis is true. It only indicates that there is no strong evidence against it in the sample you have.

The P value is often compared with a chosen significance level called alpha. If the P value is smaller than alpha, the result is called statistically significant and you reject the null hypothesis. If the P value is larger than alpha, you do not reject the null hypothesis.

The most important interpretation to avoid is the temptation to treat the P value as the probability that the null hypothesis is true. The P value looks outward from the null hypothesis toward your data, not inward from your data toward the truth.

Question-3 :- Explain the difference between a Z-test and a T-test, including when to use each ?

Answer-3 :- A Z test and a T test are two cousins in the family of hypothesis tests. They both compare a sample to some reference value, but they rely on different information about the population behind that sample.

 - A Z test assumes that the population standard deviation is known and that the sample size is large enough for the sampling distribution of the mean to behave in a clean and predictable way. When those conditions hold, the test statistic follows the standard normal distribution. Imagine measuring thousands of manufactured bolts from a factory where the variability of the production process has already been charted with precision. In that world a Z test is a natural tool.

 - A T test steps in when the population standard deviation is not known, which is the normal situation in real data analysis. You estimate the standard deviation from the sample, and that estimation introduces extra uncertainty. The test statistic then follows the Student distribution. This distribution has heavier tails that account for the added variability. When the sample size is small, this difference matters quite a lot. As the sample grows, the Student distribution slowly approaches the normal distribution and the T test behaves almost the same as the Z test.

Question-4 :- What is a confidence interval, and how does the margin of error influence its width and interpretation?

Answer-4 :- A confidence interval is a range of values that tries to capture an unknown population parameter by using information from a sample. Instead of giving a single estimate, it gives a band of plausible values. If you repeat the sampling process many times, a certain percentage of those intervals will contain the true parameter. That percentage is the confidence level. A ninety five percent confidence interval means that in the long run ninety five out of one hundred such intervals would include the true value.

The margin of error is the amount added and subtracted from the sample estimate to form this interval. It depends on the variability of the data, the sample size, and the confidence level. A large margin of error produces a wide interval. That wide interval reflects greater uncertainty and gives a less precise sense of where the true parameter lies. A small margin of error produces a narrow interval which implies more precision. You can shrink the margin of error by increasing the sample size or by choosing a lower confidence level, though lowering the confidence level trades precision for less certainty.

The interval does not claim that the true parameter moves around. The parameter is fixed. The interval moves because it is built from sample data that change from one study to the next. A confident interval is best understood as a statement about the reliability of the method rather than a probability about the parameter itself.

Question-5 :- Describe the purpose and assumptions of an ANOVA test. How does it extend hypothesis testing to more than two groups?

Answer-5 :- Analysis of Variance (ANOVA) is a statistical method used to test for significant differences between the means of three or more independent groups. It serves to determine whether at least one group mean significantly deviates from others, making it an invaluable tool for comparative analyses in various fields, such as medicine, psychology, and agricultural science.

 - Purpose: The key goal of ANOVA is to assess the null hypothesis (H0: μ1 = μ2 = ... = μk), which states that all group means are equal. By analyzing variance within and between groups, ANOVA identifies whether any observed variances are due to inherent group differences rather than random chance.
Assumptions of ANOVA:

 - Independence of Observations: The observations in each group must be independent of each other.

 - Normality: The data within each group should be normally distributed, particularly important for small sample sizes.

 - Homogeneity of Variance: The variances among the groups should be approximately equal, a condition known as homoscedasticity.


 - Extension to More Than Two Groups :-

Unlike t-tests, which can only compare means between two groups, ANOVA allows researchers to evaluate multiple groups simultaneously. By examining group means in aggregate, ANOVA controls the overall Type I error rate associated with conducting multiple pairwise t-tests.
Post-Hoc Testing: If ANOVA indicates significant differences, further post-hoc tests (e.g., Tukey’s HSD or Bonferroni correction) can be employed to pinpoint specific group differences.

 - In conclusion, ANOVA is a robust statistical method that allows for effective comparison of multiple group means while maintaining appropriate underlying assumptions to ensure the validity of results.


```python
# Question-6 :- Write a Python program to perform a one-sample Z-test and interpret the result for a given dataset ?
# answer-6 :-
import math
from scipy.stats import norm

sample = [398, 402, 405, 396, 399, 401, 403, 397, 400, 404]
population_standard_deviation = 10
population_mean = 400
sample_mean = sum(sample) / len(sample)
sample_size = len(sample)
standard_error = population_standard_deviation / math.sqrt(sample_size)
z_statistic = (sample_mean - population_mean) / standard_error
p_value = 2 * (1 - norm.cdf(abs(z_statistic)))
print("Sample mean:", sample_mean)
print("Z statistic:", z_statistic)
print("P value:", p_value)
alpha = 0.05

if p_value < alpha:
    print("The result is statistically significant. The null hypothesis is rejected.")
else:
    print("The result is not statistically significant. The null hypothesis is not rejected.")

```

```python
# Question-8 :- Generate multiple samples from a non-normal distribution and implement the Central Limit Theorem using Python ?
# Answer-8 :-
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(123)
population_size = 100000
population = np.random.exponential(scale=1.0, size=population_size)

number_of_samples = 5000

sample_size = 30

sample_means = []

for _ in range(number_of_samples):
    sample = np.random.choice(population, size=sample_size, replace=True)
    sample_mean = np.mean(sample)
    sample_means.append(sample_mean)

plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.hist(population, bins=50, edgecolor="black")
plt.title("Original Exponential Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=50, edgecolor="black")
plt.title("Distribution of Sample Means")
plt.xlabel("Sample mean")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
# Question-9 :- Write a Python function to calculate and visualize the confidence interval for a sample mean ?
# Answer-9 :-
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, t
import math

def confidence_interval(sample, confidence_level=0.95, population_standard_deviation=None):
    sample = np.array(sample)
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    if population_standard_deviation is not None:
        standard_error = population_standard_deviation / math.sqrt(sample_size)
        critical_value = norm.ppf((1 + confidence_level) / 2)
    else:
        sample_standard_deviation = np.std(sample, ddof=1)
        standard_error = sample_standard_deviation / math.sqrt(sample_size)
        critical_value = t.ppf((1 + confidence_level) / 2, df=sample_size - 1)

    margin_of_error = critical_value * standard_error

    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    print("Sample mean:", sample_mean)
    print("Confidence interval:", (lower_bound, upper_bound))

    plt.figure(figsize=(8, 4))

    plt.axvline(sample_mean, color="blue", linewidth=2, label="Sample mean")
    plt.axvline(lower_bound, color="red", linewidth=2, label="Lower bound")
    plt.axvline(upper_bound, color="green", linewidth=2, label="Upper bound")

    plt.title("Confidence Interval for the Sample Mean")
    plt.xlabel("Value")
    plt.legend()
    plt.show()

    return lower_bound, upper_bound
data = [14, 16, 15, 13, 17, 18, 16, 15, 14, 19]
confidence_interval(data, confidence_level=0.95)
```

```python
# Question-10 :- Perform a Chi-square goodness-of-fit test using Python to compare observed and expected distributions, and explain the outcome ?
# Answer-10 :- The following code performs a Chi-square goodness-of-fit test to analyze whether observed frequencies differ from expected frequencies.
import numpy as np
import scipy.stats as stats

# Define observed frequencies from a categorical variable
observed = np.array([50, 30, 20])  # Example observed counts

# Define expected frequencies assuming a uniform distribution
total_observed = np.sum(observed)
expected = np.array([total_observed / 3, total_observed / 3, total_observed / 3])  # Ensuring sum matches observed

# Perform Chi-square test
chi2_stat, p_value = stats.chisquare(observed, expected)

# Interpretation of results
alpha = 0.05  # Significance level
if p_value < alpha:
    result = "Reject the null hypothesis: Significant difference between observed and expected frequencies."
else:
    result = "Fail to reject the null hypothesis: No significant difference detected."

print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")
print("Interpretation:", result)

# Interpretation :- This Chi-square goodness-of-fit test evaluates whether the observed frequency distribution significantly deviates from the expected distribution. If the P-value is less than the significance level (e.g., 0.05), we would reject the null hypothesis, concluding that there is a significant discrepancy between the observed and expected frequencies. Conversely, a high P-value indicates no significant differences exist, affirming the validity of the expected frequencies.
```
